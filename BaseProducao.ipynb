```python
# meta de produção x produzido

meta_producao = pbi4.groupby(pbi4['ID_OP'])['QtdeTotalProduzir'].sum()
produzido = pbi5.groupby(pbi5['ID_OP'])['QtdeProduzida'].sum()
meta = pd.merge(meta_producao,produzido, on = 'ID_OP', how = 'left')
meta = meta.reset_index()
meta['MetaAtingida'] = ''
meta
```

```python
# preencher coluna MetaAtingida

#meta.mask(meta['QtdeProduzida'] < meta['QtdeTotalProduzir'], "Não")
#Sim = 1
#Não = 0
meta.loc[meta['QtdeProduzida'] <= meta['QtdeTotalProduzir'], 'MetaAtingida'] = 0
meta.loc[meta['QtdeProduzida'] >= meta['QtdeTotalProduzir'], 'MetaAtingida'] = 1
meta
```

```python
#Contagem das metas de produção

meta['MetaAtingida'].value_counts()
sim = meta['MetaAtingida'].loc[meta['MetaAtingida'] == 1].count()
nao = meta['MetaAtingida'].loc[meta['MetaAtingida'] == 0].count()
print('Meta Atingida em', sim, 'das ordens de produão')
print('Meta não Atingida em', nao, 'das ordens de produão')
```
